### Get an overview of errors

```python
# Which client_username has the most CLIENT_ERROR compared to other logging_types?
logging_type_stat = []
for client_username in client_usernames:
    df_tmp = df
    # drop rows where client_username is not X
    df_tmp_room = df_tmp[df_tmp.client_username == client_username]
    # Count how many times logging_type is CLIENT_ERROR
    loggin_types = df_tmp_room.groupby('logging_type')["client_username"].count()
    amount_of_sessions = loggin_types["CLIENT_START"]
    if "CLIENT_ERROR" in loggin_types:
        logging_type_stat.append(loggin_types['CLIENT_ERROR'])
    else:
        logging_type_stat.append(0)

    print(client_username,"Sessions:",amount_of_sessions,"Errors: ",logging_type_stat[-1], "Percentage failed: ","{:.2f}%".format(100*(logging_type_stat[-1]/amount_of_sessions))) # [-1] is the last element in the list (len(list)-1)
```

## What circuits did c4-TorScandinavia use?

```python
import json
df_tmp = df
client_username = "c4-TorScandinavia"

# drop rows where client_username is not X
df_tmp = df_tmp[df_tmp.client_username == client_username]
c4_circuits = df_tmp["latest_circuit"].dropna()
c4_circuits_unique = c4_circuits.drop_duplicates()

print("Number of circuits:",len(c4_circuits))
print("Number of unique circuits:",len(c4_circuits_unique))

# drop nan values from df_tmp["latest_circuit"]
df_tmp = df_tmp.dropna(subset=['latest_circuit'])

# parse the latest_circuit column to a list
circuits = df_tmp["latest_circuit"].tolist()
# parse the list of json strings to a list of dictionaries
circuits = [json.loads(circuit) for circuit in circuits]

# go through each circuit and create a new list with the circuit_id
circuit_list = []
for circuit in circuits:
    for c in circuit["Circuits"]:
        circuit_list.append(circuit["Circuits"][c])


circuit_list_df = pd.DataFrame({"A":circuit_list})
print("list len:",len(circuit_list_df))
circuit_list_df.drop_duplicates(inplace=True)
print("list len (dup removed):",len(circuit_list_df))
circuit_list_df
```

```python
# Count how many times each country is used as entry, middle and exit node
# TODO: This is first interesting to see, when the clients also send the circuits
# that was used in a failing session.
# See https://github.com/Master2022E/OnionRTC-experiment/issues/29#issuecomment-1356852164

#print("Len of circuit_list:",len(circuit_list))
print("Count how many times each country is used as entry, middle and exit node")
dk_counter = 0
se_counter = 0
no_counter = 0
fi_counter = 0

for circuit in circuit_list:
    entry= circuit["Entry Node"]["Country"].upper()
    middle = circuit["Middle Node"]["Country"].upper()
    exit = circuit["Exit node"]["Country"].upper()
    node_countries = [entry,middle,exit]

    for country in node_countries:
        if country == "DK":
            dk_counter += 1
        if country == "SE":
            se_counter += 1
        if country == "NO":
            no_counter += 1
        if country == "FI":
            fi_counter += 1

print("DK:",dk_counter)
print("SE:",se_counter)
print("NO:",no_counter)
print("FI:",fi_counter)
```
